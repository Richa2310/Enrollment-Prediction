```python
#Actual Enrollment vs Predicted Enrollment since 2013

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

GovtEnroll = pd.read_csv("enrollment since 2018.csv")
feature_GovtEnroll=['YEAR']
X= GovtEnroll[feature_GovtEnroll]
y= GovtEnroll.Enroll.values

PvtEnroll = pd.read_csv("Pvtenrollment since 2013.csv")
feature_PvtEnroll=['YEAR']
X1= PvtEnroll[feature_PvtEnroll]
y1= PvtEnroll.Enroll.values

TotalEnroll = pd.read_csv("Totalenrollment since 2013.csv")
feature_TotalEnroll=['YEAR']
X2= TotalEnroll[feature_TotalEnroll]
y2= TotalEnroll.Enroll.values

model=LinearRegression()
model.fit(X,y)
p=model.predict(X)

model1=LinearRegression()
model1.fit(X1,y1)
p1=model1.predict(X1)

model2=LinearRegression()
model2.fit(X2,y2)
p2=model2.predict(X2)
    
plt.figure(figsize=(8,8))
plt.scatter(X,p,c="red",s=110,label='Govt_Pred')
plt.scatter(X1,p1,c="green",s=110,label='Pvt_Pred')
plt.scatter(X2,p2,c="blue",s=110,label='All Mgt_Pred')
plt.plot(X,y,color="yellow",linewidth=5,label='Govt_Actual')
plt.plot(X1,y1,color="brown", linewidth=5,label='Pvt_Actual')
plt.plot(X2,y2,color="pink",linewidth=5,label='All mgt_Actual')
plt.legend(loc='upper right')
plt.xlabel('YEAR', fontsize=20)
plt.ylabel('ENROLLMENT', fontsize=20)
plt.title("\nBlock-Nagod, District-Satna\nActual Enrollment vs Prediction since 2013",fontweight ='bold',fontsize = 20)
plt.savefig('E:/RICHA PERSONAL/NIELET AI/Project/Richa_Project/official/enroll.jpeg')
plt.show()
```
